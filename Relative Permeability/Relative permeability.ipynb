# 2-Phase Relative Permeability

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Wyllie and Gardner Method

```python
Swc=float(input('Connate water saturation=')) #connate water saturation
```

```python
Sw=np.linspace(Swc,0.7,100)
print(Sw)
```

```python
Sew=(Sw-Swc)/(1-Swc)
```

```python
Kro=(1-Sew)**3
Krw=(Sew)**3
```

```python
plt.plot(Sw,Kro, label='Kro', linewidth=2)
plt.plot(Sw,Krw, label='Krw', linewidth=2)
plt.ylabel('Kr (Oil and Water)')
plt.xlabel('Sw')
plt.title('by Wyllie and Gardner Method')
plt.suptitle('Relative Permeability')
plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)
plt.xlim(0,1)
plt.ylim(0,1.2)
plt.legend()
plt.show()
```

### Corey's Method

```python
Sg=np.linspace(0.1,0.7,100)
print(Sg)
```

```python
Seg=Sg/(1-Swc)
```

```python
Kro=(1-Seg)**4
Krg=(Seg**3)*(2-Seg)
```

```python
plt.plot(Sg,Kro, label='Kro', linewidth=2)
plt.plot(Sg,Krg, label='Krg', linewidth=2)
plt.ylabel('Relative Permeability')
plt.xlabel('Sg')
plt.title('by Coreys Method')
plt.suptitle('Relative Permeability')
plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)
plt.xlim(0,1)
plt.ylim(0,1)
plt.legend()
plt.show()
```
